```python

# ★★★★ 구글 검색을 통해 본인의 크롬 버전에 맞는 chromedriver.exe 파일을 다운로드 ★★★★
# ★★★★ 다운로드한 chromedriver.exe을 사용할 경로에 저장 ★★★★ 

import sys    # 시스템
import os     # 시스템

import pandas as pd    # 판다스 : 데이터분석 라이브러리
import numpy as np     # 넘파이 : 숫자, 행렬 데이터 라이브러리

from bs4 import BeautifulSoup     # html 데이터 전처리
from selenium import webdriver    # 웹 브라우저 자동화
import time                       # 시간 지연
from tqdm import tqdm_notebook    # 진행상황 표시
from selenium.webdriver.common.by import By
```

```python
# CSV 파일 불러오기

keyword = pd.read_csv('C:/Users/socsoft/Downloads/internship_kwj/socsoft/법제처키워드/불공정_키워드(117) - 51-117.csv',encoding='cp949')

keyword
```

```python
# 데이터 프레임 열에 이름 없을 경우 이름달아 주기

keyword=keyword.rename(columns={'Unnamed: 0':'순번','0':'제목'})
```

```python
# 데이터 프레임에서 key 값으로 쓸 열을 리스트 형태로 가져오기
 
key = keyword["제목"]

key
```

```python
# 매크로 및 결과 저장 코드

dict = {}    # 전체 크롤링 데이터를 담을 그릇

for i in tqdm_notebook(range(0, len(key))):
    
    try:
        
    # ★★★★ 본인의 내문서 경로에 맞게 C://뒷부분을 수정 ★★★★
    # 크롬 웹브라우저 실행
        driver = webdriver.Chrome(r"C:\Users\socsoft\Downloads\internship_kwj\chromedriver.exe")

    # 사이트 주소
        driver.get("http://www.naver.com")
        time.sleep(1)


    # 검색창에 '검색어' 검색
        element = driver.find_element(By.ID, "query")
    
        element.send_keys(key[i])
        element.submit()
        time.sleep(1)

    # '뉴스' 클릭
        driver.find_element(By.LINK_TEXT, "뉴스").click()
        time.sleep(1)

    # '인링크 뉴스' 클릭
        driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li/div/div/div[1]/div[2]/a[2]').click()
        time.sleep(2) 
        
    # '인링크 뉴스 탭으로 이동'
        driver.switch_to.window(driver.window_handles[1])
        time.sleep(2) 
    
    #본문 크롤링
        
        target_info = {}  # 개별 뉴스 내용을 담을 딕셔너리 생성
        
    # 제목 크롤링
    
        overlays1=".media_end_head_headline"                        
        tit = driver.find_element(By.CSS_SELECTOR, overlays1)
        title = tit.text
      
    # 글쓴이 크롤링
        overlays2=".media_end_head_journalist_name"                                 
        nick = driver.find_element(By.CSS_SELECTOR, overlays2)
        nickname = nick.text
      
    # 날짜 크롤링
        overlays3=".media_end_head_info_datestamp_time"    
        date = driver.find_element(By.CSS_SELECTOR, overlays3)
        datetime = date.text
            
    # 내용 크롤링
        overlays4="/html/body/div/div[2]/div/div[1]/div[1]/div[2]/div[1]/div"                                 
        contents = driver.find_element(By.XPATH, overlays4)
        content=contents.text
            
    # 언론사 이름 크롤링
        overlays5="/html/body/div/div[2]/div/div[1]/div[1]/div[1]/div[1]/a/img[1]"                        
        news_company = driver.find_element(By.XPATH, overlays5)
        company = news_company.get_attribute("title")
     
    # 현재 페이지 URL 크롤링
        url = driver.current_url
        
    # 크롤링한 글은 target_info라는 딕셔너리에 담음
        target_info['제목'] = title
        target_info['기자명'] = nickname
        target_info['작성일'] = datetime
        target_info['본문'] = content
        target_info['언론사'] = company
        target_info['url'] = url
        
        
    # 각각의 글은 dict라는 딕셔너리에 담음
        dict[i] = target_info
        time.sleep(1)
        
    # 글 하나 크롤링 후 크롬 창 닫기
        driver.close()       

    # 크롤링 성공하면 글 제목을 출력
        print(i, title)        
        
    # 에러날 경우 예외 처리
    
    except:
        driver.close()
        time.sleep(2) 
        continue
      
        
print('수집한 글 갯수: ', len(dict))


# 크롤링 결과 데이터 프레임으로 만들기
result_df = pd.DataFrame.from_dict(dict, 'index')

# 데이터 프레임을 엑셀로 저장
result_df.to_excel("naver_news_content.xlsx", encoding='utf-8-sig')    
```
