```python
import openpyxl
import pandas as pd
import numpy as np
```

```python
# temp 파일
df = pd.read_excel('./TMPLAT_GTT007 (행정읍면동별 주제도).xls', header=1)
df.head()
```

```python
# 데이터 파일
df2 = pd.read_csv('./liver_cirrhosis2020.csv', encoding='cp949')
df2.head()
```

### 전처리1
- 기존 엑셀데이터 지역명에 '.' 제거
- temp 파일, 데이터 파일 '제' 글자 제거

```python
# 행정읍면동명과 지역명이 달라서 전처리 작업 필요
# 비교하는 두 파일 전부 행정지역명 동일하게 처리

# temp 파일에 있는 행정읍면동명 '제' 글자 제거
for i in range(0, len(df)):
    if '제' in df.loc[i, '행정읍면동명']:
        df.loc[i, '행정읍면동명'] = df.loc[i, '행정읍면동명'].replace('제','')

# 데이터 파일에 있는 지역명 '.', '제' 글자 제거
for i in range(0, len(df2)):
    if '.' in df2.loc[i, '지역명']:
        df2.loc[i, '지역명'] = df2.loc[i, '지역명'].replace('.','')
    elif '제' in df2.loc[i, '지역명']:
        df2.loc[i, '지역명'] = df2.loc[i, '지역명'].replace('제','')
print(df[11:13])
print(df2[30:32])
```

###  전처리2
- 동,면,읍 만 비교해서 지역명 바꾸기

```python
for i in range(0, len(df)):
    for j in range(0, len(df2)):
        try:
            # 읍/면/동이 같으면 지역명 변경
            if df.loc[i, '행정읍면동명'].split()[1] == df2.loc[j, '지역명'].split()[3]:
                df2.loc[j, '지역명'] = df.loc[i, '행정읍면동명']
            
            # 읍/면/동 앞 3글자만 같으면 지역명 변경 ex 용암1, 강서1
            elif df.loc[i, '행정읍면동명'].split()[1][:3] == df2.loc[j, '지역명'].split()[3][:3]:
                df2.loc[j, '지역명'] = df.loc[i, '행정읍면동명']
                
            # 에러 발생 시 지역명 그대로
        except:
            df2.loc[j, '지역명'] = df2.loc[j, '지역명']
```

```python
# '비표준화지표분모'에 ',' 제거해서 숫자로 변환
for i in range(0, len(df)):
    for j in range(0, len(df2)):
        if df.iloc[i][0] in df2.iloc[j][0]:
            df.loc[i, '성별'] = df2.loc[j, '성별']
            df.loc[i, '연령'] = df2.loc[j, '연령']
            df.loc[i, '비표준화지표분자'] = df2.loc[j, '비표준화지표분자']
            df.loc[i, '비표준화지표분모'] = float(df2.loc[j, '비표준화지표분모'].replace(',',''))
            df.loc[i, '비표준화지표'] = df2.loc[j, '비표준화지표지표값']
            df.loc[i, '표준화지표'] = df2.loc[j, '표준화지표지표값']
df.head()
```

### 엑셀로 추출
- xls파일(97-2003 워크시트)은 openpyxl로 읽기 불가
- xlsx파일(엑셀 통합문서)로 변환 필요

```python
# 행정읍면동명, 빈 컬럼 제거
df = df.drop(['행정읍면동명', 'Unnamed: 8'], axis=1)
df.head()
```

```python
list(df.values[0])
```

```python
# 엑셀 파일 열기
filepath = './TMPLAT_GTT007 (행정읍면동별 주제도) - 연습파일 - 복사본 - 복사본.xlsx'
wb = openpyxl.load_workbook(filepath) # Workbook 객체 생성

# 워크시트 선택
ws = wb['행정읍면동별 주제도']
ws.cell(row=1, column=1)

for rowdata in df.values:
    ws.append(list(rowdata))

    # 기존 엑셀 파일에 저장
wb.save(filepath)
wb.close()
```

```python
# 엑셀 파일 열기
filepath = './test.xlsx'
wb = openpyxl.load_workbook(filepath)

# 워크시트 선택
ws = wb['Sheet1']

# 셀마다 df 값 입력 (B3부터 차례대로 값 입력)
for i in range(len(df)):
    for j in range(len(df.loc[0])):
        ws.cell(row=3+i, column=j+2).value = df.loc[i].values[j]

# 기존 엑셀 파일에 저장
wb.save(filepath)
wb.close()
```

```python
# 엑셀 파일 열기
wb = openpyxl.load_workbook('TMPLAT_GTT007 (행정읍면동별 주제도) - 연습파일 - 복사본.xlsx')

# 워크시트 선택
ws = wb['행정읍면동별 주제도']


# 셀마다 df 값 입력 (B3부터 차례대로 값 입력)
for n, data in enumerate(df.values):
    for n2, i in enumerate(data[1:]):
        ws.cell(row=n+3, column=n2+2).value = i

# 기존 엑셀 파일에 저장
wb.save('TMPLAT_GTT007 (행정읍면동별 주제도) - 연습파일 - 복사본.xlsx')
wb.close()
```
