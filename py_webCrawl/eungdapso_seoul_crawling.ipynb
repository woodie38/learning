```python
#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import re
import math
import numpy  
import pandas as pd  
import xlwt

import random
import os   # 폴더 생성하기 위해 필요함
```

```python
#Step 2. 사용자에게 크롤링 정보를 입력 받습니다.

query_txt = '서울시응답소'

cnt = int(input('1. 크롤링 할 건수 입력: '))
page_cnt = math.ceil(cnt / 10)

f_dir = input('2. 결과 파일 저장 경로 입력(ex c:\\temp\\):')

# 저장될 파일위치와 이름지정
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir+s+'-'+query_txt)
os.chdir(f_dir+s+'-'+query_txt)

ff_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.txt'
fc_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.csv'
fx_name = f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.xls'
```

```python
#Step 3. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.

s_time = time.time()

path='C:/Users/SOCSOFT/Downloads/internship_kwj/chromedriver.exe'
driver = webdriver.Chrome(path)

driver.get('https://eungdapso.seoul.go.kr/')

time.sleep(random.randrange(2,5))

#Step 4. 시장에게 바랍니다 게시판 선택

driver.find_element_by_xpath("""//*[@id="gnb"]/div[4]/div[1]/div/ul/li[6]/a""").click()
```

```python
# 시간 계산위해 현재 타임스탬프 찍기
now1 = time.localtime()
s1 = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

print('크롤링 할 총 페이지: ', page_cnt)
print('\n')

# 게시글 요약 정보 출력

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

result_list = soup.find('div', class_ = 'pclist_table mt20').get_text()
print(result_list)
```

```python
# 현재 페이지의 내용 파일에 저장, 돌아가기로 이전 페이지 리턴

no = []    # 게시글 번호
contents = []    # 게시글 내용

n = 1

click_cnt = 1    # 현재 페이지

for x in range(1, page_cnt+1):
    print('%s 페이지 내용 수집 시작' % x)
    
    # 1페이지 게시글 10건의 상세 내역
    for i in range(2, 11):
        driver.find_element_by_xpath('//*[@id="content_cont"]/div[2]/div/form/div[4]/table/tbody/tr[%s]/td[2]/a' % i).click()
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        content = soup.find('div', 'table_inner_desc').get_text()
        print(content)
        
        no.append(n)
        contents.append(content.replace("\n","") + '\n')
        
        n += 1
        
        time.sleep(1)
        
        driver.back()
        
    time.sleep(2)
    
    if click_cnt > page_cnt+1:
        break
        
    click_cnt += 1
    
    if (click_cnt % 5 == 0):
        driver.find_element_by_xpath("""//*[@id="content_cont"]/div[2]/div/div[2]/a""").click()
    else:
        driver.find_element_by_link_text("""%s"""%click_cnt).click()
```

```python
seoul = pd.DataFrame()
seoul['번호'] = no
seoul['내용'] = contents

# txt 형식으로 저장
f = open(ff_name, 'a', encoding='UTF-8')
f.write(str(contents)+'\n')
f.close()

# csv 형식으로 저장
seoul.to_csv(fc_name, encoding='utf-8-sig')

# excel 형식으로 저장
seoul.to_excel(fx_name)

e_time = time.time()    # 검색이 종료된 시점의 타임스탬프
t_time = e_time - s_time

print('총 소요시간: %s'%round(t_time, 1))
print("파일 저장 완료: txt 파일명 : %s " %ff_name)
print("파일 저장 완료: csv 파일명 : %s " %fc_name)
print("파일 저장 완료: xls 파일명 : %s " %fx_name)

driver.close()
```
