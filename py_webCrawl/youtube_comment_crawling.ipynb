# 유튜브 댓글 수집

```python
from bs4 import BeautifulSoup
from selenium import webdriver

import time
import sys
import re
import math
import numpy as np
import pandas as pd   
import openpyxl 
import random
import os
```

```python
# 필요한 정보 입력, 저장될 파일위치와 이름

query_txt=input("1.검색키워드 입력: ")

cnt = int(input('2.크롤링할 영상 건수: '))
comm_cnt = int(input('3.각 영상에서 추출할 댓글 건수: '))

f_dir = input("4.저장할 폴더명(예:c:\\temp\\):")

now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir+s+'-'+query_txt)
os.chdir(f_dir+s+'-'+query_txt)

ff_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.txt'
fc_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.csv'
fx_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.xlsx'

s_time = time.time( )

# 영상 검색

path = 'C:/Users/SOCSOFT/Downloads/internship_kwj/chromedriver.exe'
driver = webdriver.Chrome(path)

driver.get('https://www.youtube.com')
time.sleep(5)
element = driver.find_element_by_name('search_query')

element.send_keys(query_txt)
element.submit()
time.sleep(2)

# 영상 목록 출력
# 첫 화면에 20개 나옴, 20개가 넘을 시 스크롤 후 URL 추출

def scroll_down(driver):
    driver.execute_script("window.scrollBy(0, 3000);")    # 첫 페이지 20개
    time.sleep(2)
    
if comm_cnt < 20:
    page_cnt = 1
else:
    page_cnt = math.ceil(comm_cnt/20)
    
if cnt > 20:
    i = 1
    while(i <= page_cnt):
        print('%s 회 스크롤'%i)
        scroll_down(driver)
        time.sleep(1)
        i += 1

# C:\Users\socsoft\Downloads\internship_kwj\
```

```python
# 검색된 영상의 URL 추출

time.sleep(2)

html = driver.page_source
soup1 = BeautifulSoup(html, 'html.parser')
contents = soup1.find_all('a','yt-simple-endpoint style-scope ytd-video-renderer')
count = 0
item = []


for i in contents:
    item.append(i['href'])
    count += 1
    
    if count == cnt:
        break

print('검색된 총 %s 건 동영상의 댓글 추출' % cnt)
print('\n')

# 비트맵 이미지 아이콘을 위한 대체 딕셔너리 생성
bmp_map = dict.fromkeys(range(0x10000, sys.maxunicode + 1), 0xfffd)

# 영상 댓글 출력

full_url = []
url_cnt = 0

for x in range(0, len(item)):
    url_cnt += 1
    url = 'https://www.youtube.com/' + item[x]
    full_url.append(url)
    if url_cnt == cnt:
        break
        
play_cnt = 1

for y in range(0,len(full_url)) :
    driver.get(full_url[y])
    time.sleep(2)
    
    print('\n')
    print("=" *80)
    print("%s 번째 동영상의 정보를 수집합니다." %play_cnt)

    i = 1
    while (i <= page_cnt):
        scroll_down(driver)
        time.sleep(1)
        i += 1
        
    html = driver.page_source
    soup2 = BeautifulSoup(html, 'html.parser')
    try:
        view_cnt1 = soup2.select('#count')
        view_cnt2 = view_cnt1[1].get_text().replace(',','')
        view_cnt3 = re.search('\d+', view_cnt2)
        view_cnt = int(view_cnt3.group())
    
    title1 = soup2.select('#info-contents')[0].find('h1').get_text()
    title = title1.translate(bmp_map).replace('\n', '')
    
    comm_cnt1 = soup2.select('#count')[2].get_text().replace('\n', '').replace(',','')
    comm_cnt2 = re.search('\d+', comm_cnt1)
    comm_cnt = int(comm_cnt2.group())
    
    print("%s 번째 동영상 : 조회수 %s 회, 댓글 총 %s개" %(play_cnt,view_cnt,comm_cnt))  
    print("%s 번째 동영상 제목 : %s" %(play_cnt,title))
    print("=" *80)
    print("\n")
    
# 스크롤 후 요청된 댓글 수 만큼 수집

    i = 1
    while (i <= page_cnt+1):
        scroll_down(driver)
        time.sleep(0.5)
        i += 1
        
    url_li = []
    comm_li1 = []
    comm_li2 = []
    writer_li = []
    wdate_li = []
    
    time.sleep(2)
    
    html = driver.page_source
    soup3 = BeautifulSoup(html, 'html.parser')
    
    count = 0
    
    comm_result = soup3.select('#comments > #sections > #contents')
    
    for a in range(0, comm_cnt+1):
        count += 1
        
        f = open(ff_name, 'a', encoding='UTF-8')
        f.write('\n')
        f.write('--------------------------------------------------------------------------'+'\n')
        
        # 댓글 작성id
        try:
            writer = comm_result[0].select('#author-text')[a].get_text().replace('\n','').strip()
        except IndexError:
            break
        else:
            #print('\n')
            print("%s 번째 영상의 %s 번째  댓글 " %(play_cnt,count))
            f.write("%s 번째 영상의 %s 번째  댓글 " %(play_cnt,count) + "\n")
        
        # 유튜브 URL 주소
        print("1.URL 주소:",full_url[y])
        f.write("1.유투브 URL주소: " + full_url[y] + "\n")
        url_li.append(full_url[y])
        
        # 댓글 작성id
        print("2.댓글 작성자명:", writer)
        f.write("2.댓글 작성자명: " + writer + "\n")
        writer_li.append(writer)
        
        # 댓글 내용
        comment = comm_result[0].select('#content-text')[a].get_text().replace("\n","")
        comm_li1 = comment.translate(bmp_map).replace("\n","")
        print("3.댓글 내용:", comm_li1)
        f.write("3.댓글 내용: " + comm_li1 + "\n")
        comm_li2.append(comm_li1)   
        
        f.close()
        
        if count == cnt:
            break
            
    time.sleep(2)
    play_cnt += 1
```

```python
# 저장

youtube_comm = pd.DataFrame()
youtube_comm['URL 주소']=url_li
youtube_comm['댓글작성자명']=pd.Series(writer_li)
youtube_comm['댓글 내용']=pd.Series(comm_li2)

# csv 형태로 저장하기
youtube_comm.to_csv(fc_name,encoding="utf-8-sig",index=True)

# 엑셀 형태로 저장하기
youtube_comm.to_excel(fx_name ,index=True)

e_time = time.time( )
t_time = e_time - s_time

# txt 파일에 크롤링 요약 정보 저장하기
orig_stdout = sys.stdout
f = open(ff_name, 'a',encoding='UTF-8')
sys.stdout = f

print("총 소요시간: %s 초" %round(t_time,1))
print("총 저장 건수: %s 건" %(count*cnt))

sys.stdout = orig_stdout
f.close( )

print("\n") 

print("1.요청된 총 %s 건 동영상 리뷰 중 실제 크롤링 된 리뷰수는 각 %s 건입니다" %(cnt,count))
print("2.총 소요시간: %s 초 입니다 " %round(t_time,1))
print("3.파일 저장 완료: txt 파일명 : %s " %ff_name)
print("4.파일 저장 완료: csv 파일명 : %s " %fc_name)
print("5.파일 저장 완료: xlsx 파일명 : %s " %fx_name)

driver.close( )
```
